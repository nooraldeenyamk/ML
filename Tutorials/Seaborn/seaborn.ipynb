```python
# total bild vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting a theme for the plots 
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data= tips, x = 'total_bill', y ='tip', col= 'time', hue= 'smoker', style= 'smoker', size= 'size')
```

```python
from sklearn.datasets import load_diabetes
dia = load_diabetes()

diabetes = pd.DataFrame(dia.data, columns= dia.feature_names)
```

```python
diabetes['result'] = dia.target
```

```python
diabetes.head()
```

```python
# Distribution Plot

sns.displot(diabetes['result'])
# sns.distplot(diabetes['result'])
```

```python
# Correlation : Positive and Negative
# Heat Map

correlation = diabetes.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar= True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
